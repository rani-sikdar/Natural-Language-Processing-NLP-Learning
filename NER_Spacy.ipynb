```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
nltk.download('maxent_ne_chunker')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
text = "Apple acquired Zoom in China on Wednesday 6th May 2020.\
This news has made Apple and Google stock jump by 5% on Dow Jones Index in the \
United States of America"
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
## tokenize to words
words = nltk.word_tokenize(text)
words
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
## Part of Speech tagging
pos_tags = nltk.pos_tag(words)
pos_tags
```

```python
import nltk
nltk.download('tagsets_json')
```

```python
nltk.help.upenn_tagset('IN')
```

```python
nltk.help.upenn_tagset('NNPS')
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
```

```python
chunks =nltk.ne_chunk(pos_tags,binary=True) ## NE or not a NE

for chunk in chunks:
    print(chunk)
```

```python
entities =[]
labels=[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
entities_labels = list(set(zip(entities,labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ['Entities','labels']
entities_df
```

```python
chunks =nltk.ne_chunk(pos_tags,binary=False) ## NE or not a NE
for chunk in chunks:
    print(chunk)

entities =[]
labels=[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
entities_labels = list(set(zip(entities,labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ['Entities','labels']
entities_df
```

```python
text = "Apple acquired Zoom in China on Wednesday 6th May 2020.\
This news has made Apple and Google stock jump by 5% on Dow Jones Index in the \
United States of America"
```

```python
## Load Spacy model
nlp = spacy.load('en_core_web_md')
```

```python
doc = nlp(text)
for entity in doc.ents:
  print(entity,entity.label_)
```

```python
doc = nlp(text)

entities =[]
labels =[]
position_start = []
position_end = []
pos = []
for ent in doc.ents:
    entities.append(ent)
    labels.append(ent.label_)
    position_start.append(ent.start_char)
    position_end.append(ent.end_char)
for token in doc:
    pos.append(token.pos_)
df = pd.DataFrame({'Entities':entities,'Labels':labels,
                   'Position_Start':position_start,'position_end':position_end})
df
```
